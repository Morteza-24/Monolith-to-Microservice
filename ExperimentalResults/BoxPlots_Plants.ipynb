```python
import seaborn as sns
import json
with open("HDBSCAN/HDBSCAN_Plants_output.json", "r") as f:
	hdbscan_json = json.load(f)
with open("Mono2Multi/Mono2Multi_Plants_output.json", "r") as f:
	mono2multi_json = json.load(f)
with open("Mo2oM/Mo2oM_Plants_output.json", "r") as f:
	mo2om_json = json.load(f)
```

```python
hdbscan = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in hdbscan_json:
	hdbscan["SM"].append(i["SM"])
	hdbscan["IFN"].append(i["IFN"])
	hdbscan["ICP"].append(i["ICP"])
	hdbscan["NED"].append(i["NED"])
mono2multi = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in mono2multi_json:
	mono2multi["SM"].append(i["SM"])
	mono2multi["IFN"].append(i["IFN"])
	mono2multi["ICP"].append(i["ICP"])
	mono2multi["NED"].append(i["NED"])
mo2om = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in mo2om_json:
	mo2om["SM"].append(i["SM"])
	mo2om["IFN"].append(i["IFN"])
	mo2om["ICP"].append(i["ICP"])
	mo2om["NED"].append(i["NED"])
```

```python
x = ["HDBSCAN"]*len(hdbscan["SM"]) + ["Mono2Multi"]*len(mono2multi["SM"]) + ["Mo2oM"]*len(mo2om["SM"])
y = hdbscan["SM"] + mono2multi["SM"] + mo2om["SM"]
data = {"Model": x, "SM": y}
ax = sns.boxplot(data=data, x="Model", y="SM", native_scale=True).set_title("higher is better")
```

```python
x = ["HDBSCAN"]*len(hdbscan["IFN"]) + ["Mono2Multi"]*len(mono2multi["IFN"]) + ["Mo2oM"]*len(mo2om["IFN"])
y = hdbscan["IFN"] + mono2multi["IFN"] + mo2om["IFN"]
data = {"Model": x, "IFN": y}
ax = sns.boxplot(data=data, x="Model", y="IFN", native_scale=True).set_title("lower is better")
```

```python
x = ["HDBSCAN"]*len(hdbscan["ICP"]) + ["Mono2Multi"]*len(mono2multi["ICP"]) + ["Mo2oM"]*len(mo2om["ICP"])
y = hdbscan["ICP"] + mono2multi["ICP"] + mo2om["ICP"]
data = {"Model": x, "ICP": y}
ax = sns.boxplot(data=data, x="Model", y="ICP", native_scale=True).set_title("lower is better")
```

```python
x = ["HDBSCAN"]*len(hdbscan["NED"]) + ["Mono2Multi"]*len(mono2multi["NED"]) + ["Mo2oM"]*len(mo2om["NED"])
y = hdbscan["NED"] + mono2multi["NED"] + mo2om["NED"]
data = {"Model": x, "NED": y}
ax = sns.boxplot(data=data, x="Model", y="NED", native_scale=True).set_title("lower is better")
```
