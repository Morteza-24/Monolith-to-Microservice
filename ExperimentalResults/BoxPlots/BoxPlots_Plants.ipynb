```python
import seaborn as sns
import json
import matplotlib.pyplot as plt
with open("HDBSCAN/HDBSCAN_Plants_output.json", "r") as f:
	hdbscan_json = json.load(f)
with open("Mono2Multi/Mono2Multi_Plants_output.json", "r") as f:
	mono2multi_json = json.load(f)
with open("Mo2oM/Mo2oM_Plants_output.json", "r") as f:
	mo2om_json = json.load(f)
with open("Mono2Micro/Mono2Micro_Plants_output.json", "r") as f:
	Mono2Micro_json = json.load(f)
with open("FoSCO/FoSCO_Plants_output.json", "r") as f:
	FoSCO_json = json.load(f)
with open("CoGCN/CoGCN_Plants_output.json", "r") as f:
	CoGCN_json = json.load(f)
with open("Bunch/Bunch_Plants_output.json", "r") as f:
	Bunch_json = json.load(f)
with open("MEM/MEM_Plants_output.json", "r") as f:
	MEM_json = json.load(f)
```

```python
hdbscan = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in hdbscan_json:
	hdbscan["SM"].append(i["SM"])
	hdbscan["IFN"].append(i["IFN"])
	hdbscan["ICP"].append(i["ICP"])
	hdbscan["NED"].append(i["NED"])
mono2multi = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in mono2multi_json:
	mono2multi["SM"].append(i["SM"])
	mono2multi["IFN"].append(i["IFN"])
	mono2multi["ICP"].append(i["ICP"])
	mono2multi["NED"].append(i["NED"])
mo2om = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in mo2om_json:
	mo2om["SM"].append(i["SM"])
	mo2om["IFN"].append(i["IFN"])
	mo2om["ICP"].append(i["ICP"])
	mo2om["NED"].append(i["NED"])
Mono2Micro = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in Mono2Micro_json:
	Mono2Micro["SM"].append(i["SM"])
	Mono2Micro["IFN"].append(i["IFN"])
	Mono2Micro["ICP"].append(i["ICP"])
	Mono2Micro["NED"].append(i["NED"])
FoSCO = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in FoSCO_json:
	FoSCO["SM"].append(i["SM"])
	FoSCO["IFN"].append(i["IFN"])
	FoSCO["ICP"].append(i["ICP"])
	FoSCO["NED"].append(i["NED"])
CoGCN = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in CoGCN_json:
	CoGCN["SM"].append(i["SM"])
	CoGCN["IFN"].append(i["IFN"])
	CoGCN["ICP"].append(i["ICP"])
	CoGCN["NED"].append(i["NED"])
Bunch = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in Bunch_json:
	Bunch["SM"].append(i["SM"])
	Bunch["IFN"].append(i["IFN"])
	Bunch["ICP"].append(i["ICP"])
	Bunch["NED"].append(i["NED"])
MEM = {"SM": [], "IFN": [], "ICP": [], "NED": []}
for i in MEM_json:
	MEM["SM"].append(i["SM"])
	MEM["IFN"].append(i["IFN"])
	MEM["ICP"].append(i["ICP"])
	MEM["NED"].append(i["NED"])
```

```python
colors = ['#FF6347', '#4682B4', '#32CD32', '#FFFF00', '#8A2BE2', '#DC143C', '#FF4500', '#7CFC00']

x = ["Mono2Multi"]*len(mono2multi["SM"]) +  ["Mo2oM"]*len(mo2om["SM"]) + ["HDBSCAN"]*len(hdbscan["SM"]) + ["Mono2Micro"]*len(Mono2Micro["SM"]) + ["FoSCO"]*len(FoSCO["SM"]) + ["CoGCN"]*len(CoGCN["SM"]) + ["Bunch"]*len(Bunch["SM"]) + ["MEM"]*len(MEM["SM"])
y = mono2multi["SM"] + mo2om["SM"] + hdbscan["SM"] + Mono2Micro["SM"] + FoSCO["SM"] + CoGCN["SM"] + Bunch["SM"] + MEM["SM"]
data = {"Model": x, "SM": y}

ax = sns.boxplot(data=data, x="Model", y="SM")

for i, patch in enumerate(ax.artists):
    patch.set_facecolor(colors[i])
    patch.set_edgecolor('black')

ax.set_title("higher is better")
ax.set_ylabel("SM")
plt.xticks(fontsize=8)
plt.show()
```
