```python
file_path = "Test_Projects/JPetStore/OneFileSource.java"
alpha = 1
min_samples = 2
n_clusters = 3
```

```python
from A_Hierarchical_DBSCAN_Method.main import hierarchical_DBSCAN
from A_Hierarchical_DBSCAN_Method import EvaluationMeasures as EMdbscan
from MICROscope.main import MICROscope
from MICROscope import EvaluationMeasures as EMmicroscope
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0.02,1,0.02)
```

```python
layers_DBSCAN, classes_info_DBSCAN = hierarchical_DBSCAN(file_path, alpha, min_samples, x)
```

```python
# n_clusters = list(map(lambda x: max(x)+1, layers_DBSCAN))
# n_clusters = [_ if _ > 0 else 1 for _ in n_clusters]
layers_MICROscope, classes_info_MICRO = MICROscope(file_path, alpha, [n_clusters for _ in x], x)
```

```python
ICP_MICROscope, ICP_DBSCAN = [], []
NED_MICROscope, NED_DBSCAN = [], []
for i in range(len(x)):
    ICP_DBSCAN.append(EMdbscan.ICP(layers_DBSCAN[i], classes_info_MICRO))
    ICP_MICROscope.append(EMmicroscope.ICP(layers_MICROscope[i], classes_info_DBSCAN))
    NED_DBSCAN.append(EMdbscan.NED(layers_DBSCAN[i]))
    NED_MICROscope.append(EMmicroscope.NED(layers_MICROscope[i]))
```

```python
plt.plot(x,ICP_MICROscope,'b', x,NED_MICROscope,'g', x,ICP_DBSCAN,'orange', x,NED_DBSCAN,'r')
```
