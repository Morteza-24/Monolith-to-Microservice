```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import json
import os
```

```python
results_dir = "results/Mo2oM/"
files = os.listdir(results_dir)
files = [f for f in files if f not in ["Mo2oM_JPetStore.json", "Mo2oM_DayTrader.json", "Mo2oM_AcmeAir.json", "Mo2oM_Plants.json"]]
df = pd.DataFrame()
for file in files:
	with open(results_dir + file, "rt") as f:
		data = json.load(f)
	this_df = pd.DataFrame(data).drop(columns=["microservices"])
	normalized_df = pd.DataFrame(MinMaxScaler().fit_transform(this_df.values), columns=this_df.columns)
	this_df['score'] = normalized_df["SM"] - normalized_df["ICP"] - normalized_df["IFN"] - normalized_df["NED"]
	idx = this_df["score"].argmax()
	row = this_df.iloc[[idx]].drop(columns=["n_clusters", "alpha", "threshold", "score"])
	row.insert(0, 'Dataset', file.split('.')[0][6:])
	df = pd.concat([df, row], ignore_index=True)
median_row = df.median(numeric_only=True).to_frame().T
median_row.insert(0, 'Dataset', 'Median of All Datasets')
df = pd.concat([df, median_row], ignore_index=True)
df
```

```python
print(df.to_latex(index=False, float_format="%.2f", caption="Mo2oM results on 20 projects", label="tab:consistency"))
```

```python
results_dir = "results/Mo2oM/"
files = os.listdir(results_dir)
files = [f for f in files if f not in ["Mo2oM_JPetStore.json", "Mo2oM_DayTrader.json", "Mo2oM_AcmeAir.json", "Mo2oM_Plants.json"]]
df = pd.DataFrame()
for file in files:
	with open(results_dir + file, "rt") as f:
		data = json.load(f)
	this_df = pd.DataFrame(data).drop(columns=["microservices"])
	normalized_df = pd.DataFrame(MinMaxScaler().fit_transform(this_df.values), columns=this_df.columns)
	this_df['score'] = normalized_df["SM"] - normalized_df["ICP"] - normalized_df["IFN"] - normalized_df["NED"]
	final_df = pd.DataFrame()
	for n in this_df["n_clusters"].unique():
		idx = this_df["score"][this_df["n_clusters"] == n].argmax()
		row = this_df[this_df["n_clusters"] == n].iloc[[idx]].drop(columns=["alpha", "threshold", "score"])
		final_df = pd.concat([final_df, row])
	row = final_df.median().to_frame().T.drop(columns=["n_clusters"])
	row.insert(0, 'Dataset', file.split('.')[0][6:])
	df = pd.concat([df, row], ignore_index=True)
median_row = df.median(numeric_only=True).to_frame().T
median_row.insert(0, 'Dataset', 'Median of All Datasets')
df = pd.concat([df, median_row], ignore_index=True)
df
```

```python
print(df.to_latex(index=False, float_format="%.2f", caption="Mo2oM results on 20 projects", label="tab:consistency"))
```
